```python
datasetpath = "/content/drive/MyDrive/matches_1930_2022.csv"
```

```python
Scoreabovezero = df['Score']>'0-0'
```

```python
df[Scoreabovezero]
```

```python
scorebybrazil = df['home_team'] == 'Brazil'
```

```python
df[Scoreabovezero & scorebybrazil]
```

```python
brazildf = df[Scoreabovezero & scorebybrazil].sort_values(by = 'home_score',ascending = True)
```

```python
brazildf
```

```python
brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
brazildf.plot(x ='Year' , y = 'home_score',kind='hexbin')
plt.xlabel('Year')
plt.ylabel('Home score')
plt.title('Brazil Score')
plt.show()
```

```python
brazildf.plot(x ='Year' , y = 'home_score',kind='bar')
plt.xlabel('Year')
plt.ylabel('Home score')
plt.title('Brazil Score')
plt.show()
```

```python
brazildf.plot(x ='Year' , y = 'home_score',kind='hist')
plt.xlabel('Year')
plt.ylabel('Home score')
plt.title('Brazil Score')
plt.show()
```

```python
brazildf.plot(x ='Year' , y = 'home_score',kind='scatter')
plt.xlabel('Year')
plt.ylabel('Home score')
plt.title('Brazil Score')
plt.show()
```
